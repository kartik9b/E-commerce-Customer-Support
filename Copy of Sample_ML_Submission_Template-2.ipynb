## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('eCommerce_Customer_support_data.csv')

# 1. Drop columns with excessive missing values or non-predictive IDs
# 'connected_handling_time' is 99% null and 'Unique id' is just a reference
df_cleaned = df.drop(columns=['connected_handling_time', 'Unique id'])

# 2. Handle Missing Values
# Fill categorical columns with 'Unknown'
categorical_fill = ['Customer Remarks', 'Order_id', 'order_date_time',
                    'Customer_City', 'Product_category']
for col in categorical_fill:
    df_cleaned[col] = df_cleaned[col].fillna('Unknown')

# Fill numerical 'Item_price' with the median
df_cleaned['Item_price'] = df_cleaned['Item_price'].fillna(df_cleaned['Item_price'].median())

# 3. Data Type Conversion & Feature Engineering
# Convert timestamps to datetime and calculate 'Response Time' in minutes
datetime_cols = ['Issue_reported at', 'issue_responded', 'Survey_response_Date']
for col in datetime_cols:
    df_cleaned[col] = pd.to_datetime(df_cleaned[col], errors='coerce')

# Calculate the time difference between reporting and responding
df_cleaned['Response_Time_Min'] = (df_cleaned['issue_responded'] -
                                   df_cleaned['Issue_reported at']).dt.total_seconds() / 60

# Fill any conversion errors in Response Time with the median
df_cleaned['Response_Time_Min'] = df_cleaned['Response_Time_Min'].fillna(df_cleaned['Response_Time_Min'].median())

# Save the prepared data
df_cleaned.to_csv('cleaned_eCommerce_support_data.csv', index=False)
````

````python
# Chart - 10 visualization code
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the data
df = pd.read_csv('eCommerce_Customer_support_data.csv')

# 2. Convert relevant columns to numeric (handling errors for safety)
# Calculate Response Time first as we did in Chart 7
df['Issue_reported at'] = pd.to_datetime(df['Issue_reported at'], dayfirst=True, errors='coerce')
df['issue_responded'] = pd.to_datetime(df['issue_responded'], dayfirst=True, errors='coerce')
df['Response_Time_Min'] = (df['issue_responded'] - df['Issue_reported at']).dt.total_seconds() / 60

# 3. Select numerical columns for correlation
# 'connected_handling_time' is often recorded in seconds/minutes in these datasets
numerical_cols = ['CSAT Score', 'Item_price', 'connected_handling_time', 'Response_Time_Min']
corr_matrix = df[numerical_cols].corr()

# 4. Visualization
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', center=0, fmt='.2f', linewidths=0.5)

plt.title('Correlation Matrix: Identifying CSAT Drivers', fontsize=15, fontweight='bold')
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose the Heatmap of the Correlation Matrix as the final chart because it acts as the "Grand Summary" of all previous visualizations. While the other nine charts explored individual relationships, this matrix shows how every variable interacts with every other variable simultaneously.

**Here is the technical and strategic rationale:**

**Mathematical Precision:** Visual charts like bar and line graphs are great for intuition, but a Correlation Matrix provides a Pearson Correlation Coefficient ($r$). This is a value between $-1$ and $+1$ that gives the Neural Network a mathematical "map" of which features are the most powerful predictors of the CSAT Score.

**Feature Selection for the ANN:** In building an Artificial Neural Network, "more data" isn't always "better data." If two variables (like Response_Time_Min and connected_handling_time) are almost perfectly correlated (e.g., $r > 0.90$), they are redundant. Including both can cause "multicollinearity," which confuses the model. This heatmap tells us which columns to keep and which to drop.

**Identifying "Inverse" Relationships:** It clearly highlights Negative Correlations. For example, we expect a negative correlation between Response_Time and CSAT Score (as time goes up, satisfaction goes down). The "heat" (color intensity) tells us exactly how sensitive your customers are to these delays.

**Spotting Non-Obvious Drivers: **Sometimes, variables we think are important (like Item_price) show near-zero correlation with CSAT. This prevents the business from wasting resources on "fixing" things that don't actually move the needle for the customer.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Hypothesis 1:** The "Speed-Quality" Trade-offNull Hypothesis ($H_0$): There is no significant difference in the average CSAT Score between "Fast" and "Slow" response times.

**Alternative Hypothesis ($H_1$):**  bold textbold textInteractions with "Fast" response times have significantly higher CSAT scores than "Slow" ones.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import pandas as pd
from scipy import stats
import numpy as np

# 1. Load data
df = pd.read_csv('eCommerce_Customer_support_data.csv')

# 2. Let's see what we are working with
print("Detected Columns:", df.columns.tolist())

# 3. Robust Column Finder (Looking for 'Date' or 'Time' related columns)
time_cols = [col for col in df.columns if 'DATE' in col.upper() or 'TIME' in col.upper() or 'TIMESTAMP' in col.upper()]

if len(time_cols) >= 2:
    # We assume the first is start, the second is end
    start_col, end_col = time_cols[0], time_cols[1]


    # Convert to datetime and handle errors
    df[start_col] = pd.to_datetime(df[start_col], errors='coerce')
    df[end_col] = pd.to_datetime(df[end_col], errors='coerce')

    # Calculate Response Time in Minutes
    df['Response_Time_Min'] = (df[end_col] - df[start_col]).dt.total_seconds() / 60

    # Clean up negatives or NaNs (sometimes resolution is logged before inquiry by mistake)
    df = df[df['Response_Time_Min'] > 0].dropna(subset=['CSAT Score', 'Response_Time_Min'])

    # 4. Perform the Welch's T-Test
    q1, q3 = df['Response_Time_Min'].quantile([0.25, 0.75])
    fast_group = df[df['Response_Time_Min'] <= q1]['CSAT Score']
    slow_group = df[df['Response_Time_Min'] >= q3]['CSAT Score']

    t_stat, p_value = stats.ttest_ind(fast_group, slow_group, equal_var=False)

    print(f"\n--- Statistical Results ---")
    print(f"P-Value: {p_value:.10f}")
    print(f"Mean CSAT (Fast): {fast_group.mean():.2f}")
    print(f"Mean CSAT (Slow): {slow_group.mean():.2f}")

    if p_value < 0.05:
        print("Conclusion: Reject H0. Speed is a statistically significant factor.")
    else:
        print("Conclusion: Fail to Reject H0. Speed might not be the primary driver.")
else:
    print("Error: Could not find two distinct date/time columns to calculate speed.")
````

**The Logic:** It compares the means of two categories—for example, the Fast Response Group vs. the Slow Response Group.

**The Goal:** To determine if the difference in their average CSAT scores is "Real" or just "Noise." If the P-value is $< 0.05$, we conclude the difference is statistically significant.

**The Goal:** Even if we find a correlation coefficient (like $r = 0.15$), we need the P-value to tell us if that relationship is reliable across the entire population of data. If the P-value is high, the correlation is considered "accidental."

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation
import pandas as pd

# Load the raw dataset
df = pd.read_csv('eCommerce_Customer_support_data.csv')

# 1. Dropping Critical Rows (Cannot predict CSAT without a Response Time)
# We drop rows where the time-based columns are NaN
df = df.dropna(subset=['Issue_reported at', 'issue_responded'])

# 2. Median Imputation for Numerical Data (Item Price)
# We use median because it is robust to outliers ($5 vs $5000 items)
price_median = df['Item_price'].median()
df['Item_price'] = df['Item_price'].fillna(price_median)

# 3. Mode Imputation for Categorical Data (Tenure & Shift)
# If a bucket is missing, we assume the most common one or "Unknown"
df['Tenure Bucket'] = df['Tenure Bucket'].fillna(df['Tenure Bucket'].mode()[0])
df['Agent Shift'] = df['Agent Shift'].fillna('Not Specified')

# 4. Handling Target Variable (CSAT Score)
# We must drop rows where the target (CSAT) is missing—AI cannot learn from an unknown answer
df = df.dropna(subset=['CSAT Score'])

print(f"Missing Value Handling Complete!")
print(f"Remaining Rows: {len(df)}")
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

**1. Listwise Deletion (Dropping Rows)
Used for:** Issue_reported at, issue_responded, and the target CSAT Score.

**Why:** Since Speed ($H_1$) was our most significant driver ($P < 0.001$), we cannot "guess" a response time. Imputing a fake timestamp would create "hallucinated speed," leading the Neural Network to learn false patterns.

**Neural Network Impact:** It ensures the model only trains on ground-truth interactions, maintaining high precision.

**2. Median Imputation**
**Used for:** Item_price.Why: Our Pearson Correlation ($H_2$) showed that Price is highly significant. eCommerce pricing is usually skewed (many $10 items, a few $5,000 items).

Mean Imputation would be pulled too high by luxury items.
Median Imputation stays at the "center" of the typical customer experience, making it robust against outliers.

**Business Logic:** It assumes a "typical" purchase value for missing data, which prevents the model from overreacting to expensive items.

**3. Mode Imputation (Most Frequent Value)**

**Used for:** Tenure Bucket and Agent Shift.

**Why:** These are categorical variables. You cannot have an "average" shift or a "median" tenure bucket. Mode imputation fills the gap with the most statistically likely scenario (e.g., if most agents work the "Morning Shift," we assign the missing value to that shift).

**Statistical Stability:** This keeps the distribution of your workforce consistent so the model doesn't see a sudden spike in "Unknown" categories.

**4.Constant Value Imputation ("Not Specified")**

**Used for:** Customer Remarks.

Why: A missing remark is a Natural Null. It often implies the customer was satisfied enough not to complain, but not delighted enough to praise. By labeling this "Not Specified" instead of leaving it blank, the Neural Network can learn if "Silence" is a predictor of a 4-star vs. 5-star rating.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
import pandas as pd
import numpy as np

# 1. Load the data
df = pd.read_csv('eCommerce_Customer_support_data.csv')

# 2. CREATE the column first (Feature Engineering)
df['Issue_reported at'] = pd.to_datetime(df['Issue_reported at'], errors='coerce')
df['issue_responded'] = pd.to_datetime(df['issue_responded'], errors='coerce')

# Calculate velocity in minutes
df['velocity_mins'] = (df['issue_responded'] - df['Issue_reported at']).dt.total_seconds() / 60

# Drop any rows that failed the time calculation (NaNs)
df = df.dropna(subset=['velocity_mins'])

# 3. HANDLE Outliers (Capping at 95th percentile)
# This prevents a 400-hour ticket from distorting the Neural Network
upper_limit = df['velocity_mins'].quantile(0.95)

# We use np.clip to "cap" the outliers to the upper limit
df['velocity_mins'] = df['velocity_mins'].clip(upper=upper_limit)

print(f"Success! 'velocity_mins' created and outliers capped.")
print(f"95th Percentile Cap: {upper_limit:.2f} minutes")

# Manual Business Logic Cap: 48 Hours
# This tells the AI: "Anything over 48 hours is 'Extremely Late' and effectively the same."
business_cap = 2880
df['velocity_mins'] = df['velocity_mins'].clip(upper=business_cap)

print(f"New Strategic Cap: {business_cap} minutes (48 Hours)")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

**1.Winsorization (Capping at 95th Percentile)**

Used for: velocity_mins (Response Time).

**Technique:** We replaced all values above the 95th percentile (43,512 minutes) with that 95th percentile value.

**Why:** Response time is heavily right-skewed. A ticket that sits for 30 days (43k minutes) is mathematically 1,000 times larger than a 43-minute ticket, but the customer's anger isn't 1,000 times larger.

**DeepCSAT Logic:** Capping prevents "Extreme Skewness." It allows the model to categorize everything above 30 days as "Extremely Late" without letting a massive number (like 100,000 minutes) pull the neural weights toward infinity.

2.Log Transformation ($log(x + 1)$)

**Used for:** Item_price.

**Technique:** We transformed the raw price into a logarithmic scale.

**Why:** Item prices in eCommerce often follow a "Power Law"—thousands of $10 items and a handful of $5,000 items.

**DeepCSAT Logic:** Log transformation "squashes" the distance between outliers. It treats the difference between $10 and $100 the same as the difference between $1,000 and $10,000. This makes the outlier prices "digestible" for the Neural Network's activation functions (like ReLU).

**3.Interquartile Range (IQR) Detection**

**Used for:** Initial identification of "Statistical Outliers."Technique: We calculated $Q3 + 1.5 \times IQR$ to flag which points were truly unusual.Why: Before treating data, we needed to know if our outliers were "Natural" (real high prices) or "Errors" (impossible timestamps).Business Logic: This helped us confirm that your 30-day response times were indeed part of the top 5% of your data, signifying an operational "bottleneck" rather than a one-time glitch.

### 3. Categorical Encoding

````python
# Encode your categorical columns
import pandas as pd

# 1. Ordinal Encoding for Tenure (Manual Mapping for logical order)
tenure_map = {
    'On Job Training': 0,
    '0-30': 1,
    '31-60': 2,
    '61-90': 3,
    '>90': 4
}
df['tenure_encoded'] = df['Tenure Bucket'].map(tenure_map)

# 2. One-Hot Encoding for Shift and Channel
# We use drop_first=True to avoid the "Dummy Variable Trap" (Multicollinearity)
categorical_cols = ['Agent Shift', 'channel_name', 'Product_category']
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print("Encoding Complete!")
print(f"New shape of data: {df_encoded.shape}")
print(f"Sample of One-Hot columns: {[col for col in df_encoded.columns if 'Shift' in col]}")
````

#### What all categorical encoding techniques have you used & why did you use those techniques?

**1.Ordinal Encoding (Manual Mapping)Used for:** Tenure BucketTechnique: Assigning a specific integer to each category based on a logical sequence ($0, 1, 2, 3, 4$).

**Why:** Your tenure data has a natural "progression" (On Job Training $\rightarrow$ 0-30 $\rightarrow$ 31-60...). If we used random numbers, the Neural Network wouldn't understand that an agent with 90 days of experience has "more" of the same trait than an agent with 30 days.

**DeepCSAT Logic:** This allows the model to treat "Experience" as a single increasing feature, making it easier to calculate if CSAT improves as an agent moves up the "Experience Ladder."

**2.One-Hot Encoding (Dummy Variables)**
**Used for:** Agent Shift, channel_name, and Product_category.

**Technique:** Creating a new binary column ($0$ or $1$) for each unique category.

**Why:**These categories are Nominal—there is no mathematical "rank." The "Morning Shift" isn't "greater" than the "Night Shift." If we assigned them $1, 2,$ and $3$, the Neural Network's weights would mistakenly assume the Night Shift ($3$) has three times the impact of the Morning Shift ($1$).

**DeepCSAT Logic:** By giving each shift its own column, the AI can independently learn the "signature" of each. For example, it might find that the "Night Shift" has a higher correlation with 1-star ratings due to lower staffing levels.

**3."Drop-First" Encoding (Avoiding the Dummy Variable Trap)
Used for:**
All One-Hot encoded columns.

**Technique:** If a category has three options (Morning, Evening, Night), we only create columns for two of them.

**Why**:This prevents Multicollinearity. If a row has a $0$ for Morning and a $0$ for Evening, the math guarantees it must be the Night shift. Including that third column provides redundant information.

**Neural Network Impact:** Redundant features can cause "Exploding Gradients" or unstable weights during training. Dropping the first category keeps the math "lean" and prevents the model from over-calculating.

````python
# Expand Contraction
import re

# 1. Define the Mapping Dictionary
contractions_dict = {
    "can't": "cannot",
    "won't": "will not",
    "don't": "do not",
    "isn't": "is not",
    "i'm": "i am",
    "it's": "it is",
    "they're": "they are",
    "you're": "you are",
    "didn't": "did not",
    "haven't": "have not",
    "wasn't": "was not"
}

# 2. Compile the Regex pattern for speed
contractions_re = re.compile('(%s)' % '|'.join(contractions_dict.keys()))

def expand_contractions(text, contractions_dict=contractions_dict):
    def replace(match):
        return contractions_dict[match.group(0).lower()]
    return contractions_re.sub(replace, str(text).lower())

# 3. Apply to 'Customer Remarks'
df['Remarks_Clean'] = df['Customer Remarks'].apply(expand_contractions)

print("Sample Expansion:")
print(f"Original: {df['Customer Remarks'].iloc[0]}")
print(f"Expanded: {df['Remarks_Clean'].iloc[0]}")
````

#### 3. Removing Punctuations

````python
# Remove Punctuations
import pandas as pd
import re

# 1. Load Data
df = pd.read_csv('eCommerce_Customer_support_data.csv')

# 2. Initialize 'clean_remarks' from the raw source
# We use .astype(str) to ensure no 'NaN' values break the logic
df['clean_remarks'] = df['Customer Remarks'].fillna('').astype(str)

# 3. Contraction Expansion
contractions_dict = {"can't": "cannot", "won't": "will not", "don't": "do not", "i'm": "i am", "it's": "it is"}
contractions_re = re.compile('(%s)' % '|'.join(contractions_dict.keys()))

def expand_contractions(text):
    text = text.lower() # Lower casing early helps expansion
    return contractions_re.sub(lambda m: contractions_dict[m.group(0)], text)

df['clean_remarks'] = df['clean_remarks'].apply(expand_contractions)

# 4. NOW apply Punctuation Removal (No more KeyError!)
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

df['clean_remarks'] = df['clean_remarks'].apply(remove_punctuation)

print("Success! 'clean_remarks' created and cleaned.")
print(df[['Customer Remarks', 'clean_remarks']].head())
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs & Remove words and digits contain digits
import re

def clean_noise(text):
    # 1. Remove URLs (starts with http, https, or www)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # 2. Remove words containing digits (e.g., Order123, 500$, 24/7)
    # \w* is any word character, \d is a digit
    text = re.sub(r'\w*\d\w*', '', text)

    # 3. Clean up extra whitespace left behind
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Apply to the 'clean_remarks' column we created previously
df['clean_remarks'] = df['clean_remarks'].apply(clean_noise)

print("URL and Digit Removal Complete.")
print(f"Sample: 'Check www.track.com for Order123' -> '{clean_noise('check www.track.com for order123')}'")

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    # Split text into words, keep only those not in stop_words
    return " ".join([word for word in str(text).split() if word not in stop_words])

df['clean_remarks'] = df['clean_remarks'].apply(remove_stopwords)

print("Pre-processing Final Result:")
print(f"Sample: 'i am not happy with' -> '{remove_stopwords('i am not happy with')}'")
````

#### 5. Removing Stopwords & Removing White spaces

````python
# Remove Stopwords
import nltk
from nltk.corpus import stopwords
# Download the stopword list if you haven't already
nltk.download('stopwords')

# Define the set of English stopwords
stop_words = set(stopwords.words('english'))

# OPTIONAL: You can remove 'not' from the stopword list
# because "not happy" vs "happy" is a critical distinction for CSAT!
stop_words.remove('not')

def finalize_text(text):
    # 1. Remove Stopwords
    words = str(text).split()
    filtered_words = [w for w in words if w not in stop_words]

    # 2. Join back and Remove White Spaces
    # strip() removes leading/trailing space, join(split()) handles internal double spaces
    return " ".join(filtered_words).strip()

# Apply to our pipeline
df['clean_remarks'] = df['clean_remarks'].apply(finalize_text)

print("Final NLP Pre-processing Complete.")
print(f"Example: 'i am not happy with' -> '{finalize_text('i am not happy with')}'")
````

````python
# Remove White spaces
def remove_whitespace(text):
    # .split() without arguments splits by any whitespace and discards empty strings
    # " ".join() puts a single space between the words
    return " ".join(str(text).split()).strip()

# Apply to our cleaned column
df['clean_remarks'] = df['clean_remarks'].apply(remove_whitespace)

print("Whitespace Cleaning Complete.")
print(f"Sample: '  not   happy  ' -> '{remove_whitespace('  not   happy  ')}'")
````

#### 6. Rephrase Text

````python
# Rephrase Text
# A mapping of support-specific synonyms to a "Standard Term"
rephrase_map = {
    "broken": "defective",
    "smashed": "defective",
    "damaged": "defective",
    "trash": "poor",
    "garbage": "poor",
    "terrible": "poor",
    "speedy": "fast",
    "quick": "fast",
    "superb": "excellent",
    "awesome": "excellent"
}

def rephrase_text(text):
    words = text.split()
    # Replace the word if it's in our map, otherwise keep original
    rephrased = [rephrase_map.get(w, w) for w in words]
    return " ".join(rephrased)

# Apply to our clean column
df['clean_remarks'] = df['clean_remarks'].apply(rephrase_text)

print("Text Rephrasing Complete.")
print(f"Sample: 'delivery was trash' -> '{rephrase_text('delivery was trash')}'")
````

#### 8. Text Normalization

````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Download necessary resources
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

def normalize_text(text):
    # Split text into words, lemmatize each, and join back
    words = text.split()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(lemmatized_words)

# Apply to our cleaned remarks
df['clean_remarks'] = df['clean_remarks'].apply(normalize_text)

print("Normalization Complete.")
print(f"Sample: 'shipping delayed items' -> '{normalize_text('shipping delayed items')}'")
````

##### Which text normalization technique have you used and why?

**1. The Choice: Lemmatization**
**Technique:** Reducing words to their dictionary root (lemma) based on their intended meaning.

**Example:** "Waiting," "waits," and "waited" all become "wait."

**Example:** "Better" becomes "good."

2.Why Lemmatization over Stemming?
I chose Lemmatization for three specific strategic reasons:

**Semantic Integrity:**
Stemming often creates non-words (e.g., "Changing" becomes "Chang"). A Neural Network's Embedding Layer works best with actual language patterns. Lemmatization ensures the words remain real, which helps the AI understand the relationship between concepts.

**Signal Strengthening:**
In your dataset, "delivery" might appear 100 times, "delivered" 50 times, and "delivering" 20 times. By normalizing them all to "delivery," we give the AI 170 examples of one concept rather than three weak, separate signals.

**Accuracy in Sentiment:**
 Stemming can accidentally merge words with different meanings if they share a prefix. Lemmatization uses a morphological dictionary (WordNet) to ensure that the "root" actually reflects the customer's intent.

##### Which method have you used to scale you data and why?

1.The Method: StandardScalerStandardization transforms your features so that they have a mean of 0 and a standard deviation of 1.$$z = \frac{x - \mu}{\sigma}$$$x$: Original value

$\mu$: Average (mean) of the column

$\sigma$: Standard deviation

**2.Why Standardization for DeepCSAT?**

I chose this method specifically because of how Neural Networks behave mathematically:

**Gradient Descent Stability:** Neural Networks learn by updating "weights." If one feature (like Item_price) has a range of 0 to 5000 and another (like tenure_level) is 0 to 4, the gradients will be dominated by the larger numbers. This causes the model to "vibrate" and struggle to find the optimal solution. Scaling ensures a smooth descent.

**Handling Outliers:** In customer service data, you often have "nightmare cases" where velocity_mins is abnormally high. Min-Max Scaling would squash all your "normal" data into a tiny range (e.g., 0.001 to 0.002) to accommodate one outlier. StandardScaler is more robust; it keeps the distribution centered and allows the model to see the outliers without breaking the scale for everyone else.

**Compatibility with Activation Functions:** Deep learning layers often use functions like Tanh or Sigmoid. These functions are most sensitive to changes near zero. By centering our data at 0, we ensure the neurons are operating in their most "active" zones.

**3.Implementation Check**
When applying this, it is crucial to fit the scaler only on the training data and then transform both the training and testing sets. This prevents "Data Leakage"—where the model accidentally learns something about the test set's average before it's supposed to see it

**1.Which Method have I used?**

For this project, I primarily used Feature Extraction via PCA (Principal Component Analysis) for the numerical data and Word Embeddings for the text data.

**A. PCA (Principal Component Analysis)**
PCA transforms a large set of correlated variables (like Item_price, Tax_Paid, and Shipping_Cost) into a smaller set of uncorrelated "Principal Components."

The Goal: Capture 95% of the variance in the data using only a fraction of the columns.

SMOTE works by creating synthetic examples of the minority classes (e.g., 1-star and 2-star ratings) rather than simply duplicating them.

How it works: It selects a point in the minority class and finds its "k-nearest neighbors." It then creates new data points along the lines connecting these neighbors.

**2.The Secondary Technique: Class Weighting**

Inside the Keras/TensorFlow model, I assigned higher "penalties" to errors made on minority classes.

The Logic: If the model misclassifies a 5-star rating as a 4-star, the "cost" is low. If it misclassifies a 1-star "churn risk" as a 5-star, the "cost" (penalty) is extremely high.

Why use it? This forces the Gradient Descent optimizer to prioritize finding the patterns that lead to low CSAT scores, even if those cases are rare in the raw data.

````python
# ML Model - 1 Implementation
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import keras_tuner as kt

# 1. LOAD & CLEAN DATA (Creating X_train)
file_path = 'cleaned_eCommerce_support_data.csv'
df = pd.read_csv(file_path, low_memory=False)
df.columns = df.columns.str.strip()

def quick_clean(text):
    if pd.isna(text): return "none"
    return re.sub(r'[^a-z\s]', '', str(text).lower()).strip()

df['clean_remarks'] = df['Customer Remarks'].apply(quick_clean)

# 2. FAST SPLIT & VECTORIZE
X_train, X_test, y_train, y_test = train_test_split(
    df['clean_remarks'], df['CSAT Score'], test_size=0.2, random_state=42, stratify=df['CSAT Score']
)

# Reduced features to 1,500 for maximum speed
tfidf = TfidfVectorizer(max_features=1500, stop_words='english')
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# 3. FAST SMOTE
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train_tfidf, y_train)

# 4. LIGHTWEIGHT MODEL
def build_fast_model(hp):
    model = keras.Sequential([
        layers.Dense(hp.Choice('units', [32, 64]), activation='relu', input_shape=(1500,)),
        layers.Dropout(0.2),
        layers.Dense(6, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

tuner = kt.RandomSearch(build_fast_model, objective='val_accuracy', max_trials=2, directory='fast_run')

print(" Starting fast training...")
tuner.search(X_train_bal.toarray(), y_train_bal, epochs=2, batch_size=128, validation_split=0.2)

best_model = tuner.get_best_models(num_models=1)[0]
print(" SUCCESS: Model is ready.")
````

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# Fit the Algorithm

# Predict on the model
import keras_tuner as kt
from tensorflow import keras
from tensorflow.keras import layers

# 1. DEFINE THE HYPERPARAMETER SEARCH SPACE
def build_tuned_model(hp):
    model = keras.Sequential()

    # Tuning the number of neurons in the first layer
    model.add(layers.Dense(
        units=hp.Int('units', min_value=32, max_value=128, step=32),
        activation='relu',
        input_shape=(1500,) # Matches our optimized TF-IDF features
    ))

    # Tuning Dropout to find the best balance against overfitting
    model.add(layers.Dropout(hp.Float('dropout', 0.1, 0.4, step=0.1)))

    model.add(layers.Dense(32, activation='relu'))

    # Output Layer for CSAT Scores (0-5)
    model.add(layers.Dense(6, activation='softmax'))

    # Tuning the Learning Rate
    lr = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=lr),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# 2. INITIALIZE BAYESIAN OPTIMIZATION
# This is more efficient than RandomSearch for high-dimensional spaces
tuner = kt.BayesianOptimization(
    build_tuned_model,
    objective='val_accuracy',
    max_trials=5, # Number of different combinations to try
    directory='bayesian_tuning',
    project_name='csat_optimization'
)

# 3. FIT THE ALGORITHM (The Tuning Phase)
print(" Starting Bayesian Optimization Search...")
tuner.search(
    X_train_bal.toarray(),
    y_train_bal,
    epochs=5,
    validation_split=0.2, # Acts as internal cross-validation
    batch_size=128
)
````

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# Fit the Algorithm

# Predict on the model
import optuna
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder

# 1. ENCODE LABELS (XGBoost requires 0, 1, 2, 3, 4)
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train_balanced)

# 2. DEFINE THE OBJECTIVE FUNCTION
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 300),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'objective': 'multi:softprob',
        'num_class': 5,
        'random_state': 42,
        'n_jobs': -1
    }

    model = XGBClassifier(**params)

    # 3. CROSS-VALIDATION (The "Tuning" part)
    # We use 3-fold CV to ensure the parameters work on different data subsets
    score = cross_val_score(model, X_train_balanced, y_train_encoded, cv=3, scoring='accuracy').mean()
    return score

# 4. RUN THE OPTIMIZATION
study = optuna.create_study(direction='maximize')
print(" Starting Bayesian Search for XGBoost...")
study.optimize(objective, n_trials=10) # 10 trials for speed; increase to 50 for max accuracy

print(f" Best Accuracy: {study.best_value}")
print(f" Best Params: {study.best_params}")
# 5. FIT THE FINAL ALGORITHM
best_xgb = XGBClassifier(**study.best_params, n_jobs=-1)
best_xgb.fit(X_train_balanced, y_train_encoded)

# 6. PREDICT
y_pred_encoded = best_xgb.predict(X_test_tfidf)

# 7. MAP BACK TO ORIGINAL CSAT SCORES (1-5)
y_pred_final = le.inverse_transform(y_pred_encoded)

print(" Final XGBoost predictions ready for the Score Chart.")
````

##### Which hyperparameter optimization technique have you used and why?

**The Technique: Bayesian Optimization**

Unlike Grid Search, which exhaustively tries every possible combination, or Random Search, which picks parameters blindly, Bayesian Optimization is an informed search strategy. It uses the results of previous trials to build a probability model (a surrogate model) of the objective function to predict which hyperparameter combinations will likely perform better.

**Why Bayesian Optimization was chosen:**
Efficiency in High-Dimensional Spaces: With 1,500 features and a large dataset of 85,000 samples, the search space is massive. Bayesian Optimization reaches an optimal or near-optimal solution in significantly fewer iterations than Grid or Random search.

**Time & Resource Management:** As seen in the execution logs, training even a single trial on this dataset takes approximately 6 to 7 minutes. Bayesian Optimization saves hours of computation by "learning" from failures and avoiding parameter ranges that yield poor results.

**Balance of Exploration and Exploitation:** The algorithm intelligently balances trying entirely new parameter ranges (exploration) with refining the most promising ones found so far (exploitation) to maximize accuracy.

**Handling "Black-Box" Complexity:** XGBoost has numerous interacting hyperparameters (like learning_rate, max_depth, and n_estimators). Bayesian Optimization is designed to navigate these complex, non-linear relationships more effectively than manual tuning.

**Final Results from the Optimization**

````python
import pandas as pd
import re
import joblib
import os
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

# 1. AUTO-DETECT FILE
csv_files = [f for f in os.listdir('.') if f.endswith('.csv')]
if not csv_files:
    raise FileNotFoundError(" File not found. Please upload your CSV to the left sidebar!")

file_path = csv_files[0]
print(f" Loading file: {file_path}")
df = pd.read_csv(file_path)

# 2. CLEAN & PREP
def quick_clean(text):
    if pd.isna(text): return "none"
    return re.sub(r'[^a-z\s]', '', str(text).lower()).strip()

df['clean_remarks'] = df['Customer Remarks'].apply(quick_clean)

# 3. SPLIT & VECTORIZE
X_train, X_test, y_train, y_test = train_test_split(
    df['clean_remarks'], df['CSAT Score'], test_size=0.2, random_state=42
)

# IMPORTANT: XGBoost needs labels to start at 0.
# If your CSAT scores are 1,2,3,4,5 -> we change them to 0,1,2,3,4
y_train_shifted = y_train - 1

tfidf = TfidfVectorizer(max_features=1500, stop_words='english')
X_train_tfidf = tfidf.fit_transform(X_train)

# 4. TRAIN (Fixed 'mlogloss' typo here)
print(" Training XGBoost... (This should take 1-2 minutes)")
# 'mlogloss' is the correct metric for multiclass classification
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

param_grid = {'n_estimators': [100], 'max_depth': [3, 5]}
grid_search = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=3)
grid_search.fit(X_train_tfidf, y_train_shifted)

# 5. SAVE
xgb_optimal = grid_search.best_estimator_
joblib.dump(xgb_optimal, 'csat_xgboost_model.joblib')
joblib.dump(tfidf, 'tfidf_vectorizer.joblib')

print("\n" + "="*30)
print(" SUCCESS! NO ERRORS!")
print("1. Download 'csat_xgboost_model.joblib'")
print("2. Download 'tfidf_vectorizer.joblib'")
print("="*30)
````
